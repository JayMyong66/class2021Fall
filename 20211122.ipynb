```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd  # 플레이할때 쓰는 것.
```

```python
def phasor_sine(amp, sr, dur, freq, phase): # 위 phasor 과정을 압축해서 함수로 만듦.
  t = np.arange(1, sr * dur+1)/sr           # 입력하는 것이 아닌 계산되는 time, theta, signal 값들.
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```
